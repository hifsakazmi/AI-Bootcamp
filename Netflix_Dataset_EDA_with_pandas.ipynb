**Question 12:**
Find all TV shows added after January 1, 2020, that were released in India.

```python
import datetime as dt
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')

df[(df['country'].astype(str).str.contains('India')) &
 (df['date_added'] > dt.datetime(2020, 1, 1))]
```

**Question 13:**
